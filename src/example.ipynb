```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

from shapely import wkb
# this is used to cache function return values
# here specifically: SQL query text --> return value
from joblib import Memory 

# custom code
from utilities import run_sql
```

```python
# Query the database to receive the outlines of administrative regions in Germany. (Europe available!)
df_administrative = run_sql('administrative_boundaries', params={'filter': 'DE21%%'}) # double percent gets interpreted as single percent in SQL Query

# Convert to GeoDataFrame
gdf_administrative = gpd.GeoDataFrame(df_administrative, geometry=df_administrative.geom.apply(wkb.loads), crs=3035)

# Transform to Web Mercator Projection, set index and drop binary column (not needed anymore)
gdf_administrative = gdf_administrative.to_crs(3857).set_index('nuts_id').drop(columns=['geom'])

# Plot border outlines
gdf_administrative.boundary.plot()

# Save results
plt.savefig('../figures/Oberbayern.png')
gdf_administrative.to_csv('../results/Oberbayern.csv')
```
