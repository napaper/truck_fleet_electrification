```python
# Configure data file paths for CSV loading
# All data files are assumed to be in the input directory structure
# This notebook is executed from the data_handling directory

# Use relative paths from data_handling directory
input_base_path = "../input"

# Define paths to existing CSV data files
fleet_csv_path = os.path.join(input_base_path, "home", "fleet.csv")
tracks_csv_path = os.path.join(input_base_path, "stations", "tracks.csv")
track_ids_csv_path = os.path.join(input_base_path, "stations", "track_ids.csv")

# Debug: Print file paths
print(f"Current working directory: {os.getcwd()}")
print(f"Input base path: {input_base_path}")
print(f"Fleet CSV path: {fleet_csv_path}")

# Verify that all required CSV files exist
required_files = [fleet_csv_path, tracks_csv_path, track_ids_csv_path]
for file_path in required_files:
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        raise FileNotFoundError(f"Required data file not found: {file_path}")

print("All required data files found successfully.")
```

```python
# Load trip data from existing CSV file
# This replaces the previous database query with direct CSV loading
df_trips = pd.read_csv(tracks_csv_path, index_col='track_id')

# Display basic information about the loaded trip data
print(f"Loaded {len(df_trips)} trip records")
print(f"Columns: {list(df_trips.columns)}")
print(f"Data types:\n{df_trips.dtypes}")

# Display the first few rows for verification
df_trips.head()
```

```python
# Load fleet information from existing CSV file
df_fleet = pd.read_csv(fleet_csv_path, index_col='vehicle_id')

# Create mapping dictionary from vehicle ID to freight forwarder ID
vehicle_to_ff = df_fleet['fleet_test_id'].to_dict()

# Add freight forwarder information to trips dataframe
df_trips['freight_forwarder'] = df_trips['vehicle_id'].map(vehicle_to_ff)

# Export updated data to CSV files for further analysis
# Note: These operations now work with local data instead of database queries
df_trips.to_csv(tracks_csv_path)
df_fleet.to_csv(fleet_csv_path)

# Display fleet information for verification
print(f"Loaded {len(df_fleet)} fleet records")
df_fleet.head()
```

```python
# Load track IDs data from existing CSV file
# This section handles the track ID mapping data and ensures data integrity
df_track_ids = pd.read_csv(track_ids_csv_path)

# Validate that the index matches the track_id_new column before processing
# This maintains the same validation logic as the original code
assert all(df_track_ids.index == df_track_ids['track_id_new']), 'Index does not match track_id_new'

# Set the track_id_new column as the index for proper data structure
df_track_ids.set_index('track_id_new', inplace=True)

# Export processed track IDs data for further analysis
df_track_ids.to_csv(track_ids_csv_path)

# Display track IDs data for verification
print(f"Loaded {len(df_track_ids)} track ID mappings")
df_track_ids.head()
```

```python
# Perform data quality checks and provide summary statistics
# This new cell helps verify the loaded data integrity

print("=== DATA LOADING SUMMARY ===")
print(f"Trips loaded: {len(df_trips)}")
print(f"Fleet records loaded: {len(df_fleet)}")
print(f"Track ID mappings loaded: {len(df_track_ids)}")

# Check for missing values in key columns
print("\n=== MISSING VALUES CHECK ===")
print("Trips missing values:")
print(df_trips.isnull().sum())

print("\nFleet missing values:")
print(df_fleet.isnull().sum())

# Verify data consistency
print("\n=== DATA CONSISTENCY CHECK ===")
print(f"Unique vehicles in trips: {df_trips['vehicle_id'].nunique()}")
print(f"Unique vehicles in fleet: {df_fleet.index.nunique()}")
print(f"Trips with freight forwarder mapping: {df_trips['freight_forwarder'].notna().sum()}")

print("\nData loading and processing completed successfully!")
```
